```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,50)
y = 2*x+1

plt.figure()
plt.plot(x,y)
```

```python
y2 = x**2
plt.figure()

l1, = plt.plot(x,y,color='red',linewidth=2,linestyle='--')
l2, = plt.plot(x,y2)

plt.xlim(-1,0.5)
plt.ylim(0,2)

plt.xlabel('I am X')
plt.ylabel('I am Y')

plt.legend(handles=[l1,l2,],labels=['l1','l2'],loc='best')

plt.show()
```
