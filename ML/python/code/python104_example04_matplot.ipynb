```python
## http://matplotlib.org/gallery/index.html

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,50)
y = 2*x + 1

plt.figure(num=1, figsize=(8,5))
plt.plot(x,y)

ax = plt.gca()
ax.spines['right'].set_color('red')
ax.spines['top'].set_color('blue')
ax.spines['left'].set_color('g')
ax.spines['bottom'].set_color('k')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

#plt.show()
```

```python
x0 = 1
y0 = 2*x0 + 1

plt.scatter(x0, y0, s=50, color='b')
plt.plot([x0,x0], [y0,0], 'k--', lw=2.5)
plt.plot([0,x0], [y0,y0], 'g--', lw=2.5)

plt.annotate(r'$2x+1=%s$' % y0, xy=(x0,y0), xycoords='data', xytext=(+30,-30), textcoords='offset points',
            fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

plt.text(-3.7,3,r'$This\ is\ the\ some\ text,\mu\ \sigma_i\ \alpha_t$', fontdict={'size':16,'color':'r'})

plt.show()
```

```python
## 散点图

n=1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X) # for color value

plt.scatter(X, Y, s=75, c=T, alpha=0.5)
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))

plt.show()
```

```python
## 柱状图

n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1, facecolor='#9999ff')
plt.bar(X,-Y2)

for x,y in zip(X,Y1):
    plt.text(x+0.0,y+0.05, '%.2f' % y, ha='center', va='bottom')
    
for x,y in zip(X,-Y2):
    plt.text(x+0.0,y-0.05, '%.2f' % y, ha='center', va='top')

plt.show()
```

```python
## 等高线图

def f(x,y): # the height function
    return (1-x/2 + x**5 + y**3)*np.exp(-x**2-y**2)

n=256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X,Y,f(X,Y),15,alpha=0.75,cmap=plt.cm.hot)

# 等高线
C = plt.contour(X,Y,f(X,Y),8,colors='black', linewidth=0.5)

plt.clabel(C,inline=True,fontsize=10)

plt.show()
```

```python
## 图像

## 3D
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow')

ax.set_zlim(-2,2)

plt.show()
```
