## 定义函数

```python
# 返回一个值
def my_abs(x):
    if not isinstance(x,(int,float)):
        raise TypeError('bad type error')
    if x>=0:
        return x;
    else:
        return -x;
```

```python
my_abs(-10)
```

```python
my_abs('A')
```

```python
# 返回多个值,返回值是tuple
def move(x,y,step):
    nx = x+step
    ny = y-step
    return nx,ny
```

```python
x,y = move(100,200,10)
x,y
```
